```python
def handle_outliers(data):
   num_columns = []
   for i in list(data.columns):
       if data[i].dtype == "int64" or data[i].dtype == "float64":
        num_columns.append(i)
   for column_name in num_columns:
      Q1 = data[column_name].quantile(0.25)
      Q3 = data[column_name].quantile(0.75)
      IQR = Q3 - Q1
      upper_limit = Q3 + 1.5 * IQR
      lower_limit = Q1 - 1.5 * IQR
      data[column_name] = data[column_name].clip(lower=lower_limit, upper=upper_limit)
   return data

# Assuming dd is your DataFrame
dd = handle_outliers(dd)
```

##**TRAIN TEAST SPLIT**


```python
X = dd.drop("Outcome", axis=1)
Y = dd["Outcome"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=10)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

##**MODELING**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
import joblib
# Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
# Evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import confusion_matrix
```

#**LOGISTIC REGRESSION**

```python
# Hyperparameter tuning
log_reg = LogisticRegression(random_state=49)
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}
grid_search = GridSearchCV(estimator=log_reg, param_grid=param_grid, cv=5)
grid_search.fit(X_train, Y_train)
best_grid = grid_search.best_params_
print("Best Grid:", best_grid)

# Logistic Regression with the best parameters
best_C = best_grid['C']
best_penalty = best_grid['penalty']
best_solver = best_grid['solver']

# Training a Model
model = LogisticRegression(C=best_C, penalty=best_penalty, solver=best_solver, random_state=49)
model.fit(X_train, Y_train)

# Prediction
log_test_prediction = model.predict(X_test)
log_train_prediction = model.predict(X_train)

# Accuracy
test_accuracy = accuracy_score(Y_test, log_test_prediction)
train_accuracy = accuracy_score(Y_train, log_train_prediction)
LR_CV = cross_val_score(model, X_test, Y_test, cv=5, scoring="accuracy").mean()

# Evaluation metrics
print("Model_Train_Accuracy :", train_accuracy)
print("Model_Test_Accuracy :", test_accuracy)
print("Cross Validation Score :", LR_CV)
```

#**K NEAREST NEIGHBORS**

```python
# Hyperparameter tuning for KNN
knn_model = KNeighborsClassifier()
param_grid = {'n_neighbors': list(range(1, 71))}
cv_classifier = GridSearchCV(estimator=knn_model, param_grid=param_grid, cv=5, scoring="accuracy")
cv_classifier.fit(X_train, Y_train)
best_param = cv_classifier.best_params_
print("Best Param: ", best_param)

# Extract the best number of neighbors
best_neighbor = best_param['n_neighbors']

# KNN Model with the best parameters
knn_model = KNeighborsClassifier(n_neighbors=best_neighbor)
knn_model.fit(X_train, Y_train)

# Prediction
knn_test_prediction = knn_model.predict(X_test)
knn_train_prediction = knn_model.predict(X_train)

# Accuracy
knn_test_accuracy = accuracy_score(Y_test, knn_test_prediction)
knn_train_accuracy = accuracy_score(Y_train, knn_train_prediction)
KNN_CV = cross_val_score(knn_model, X_test, Y_test, cv=5, scoring="accuracy").mean()

# Evaluation metrics
print("Model_Train_Accuracy :", knn_train_accuracy)
print("Model_Test_Accuracy :", knn_test_accuracy)
print("Cross Validation Score :", KNN_CV)
```

#**ROC CURVE**

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Get predicted probabilities for the positive class
log_test_probabilities = model.predict_proba(X_test)[:, 1]  # Assuming 'model' is your Logistic Regression model

# Calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(Y_test, log_test_probabilities)

# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc_score(Y_test, log_test_probabilities):.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Logistic Regression)', fontsize=30)
plt.legend(loc='lower right')
plt.show()
```
